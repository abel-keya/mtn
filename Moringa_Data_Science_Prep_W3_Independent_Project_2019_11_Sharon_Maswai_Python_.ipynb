# Merging and Concatenating Dataframes

```
#merging the df_geo dataset and telecom dataset at the cell_id column using inner join.
df_joint=telecom.merge(df_geo, how='inner', on='cell_id')
df_joint.head()
```

```
#The number of columns and rows in the table where the cell_id data is the sam in the two tables.
df_joint.shape
```

```
#Merging telecom2 and df_geo dataset at the cell_id column which is common in teh two dtaframes. 
#Inner join is used to ensure that only rows with similar cell_id are merged.
df_joint2=telecom2.merge(df_geo, how='inner', on='cell_id')
df_joint2.head()
```

```
#Number or rows and columns in the newly created dataframe
df_joint2.shape
```

```
#Merging telecom3 and df_geo dataset at the cell_id column which is common in teh two dtaframes. 
df_joint3=telecom3.merge(df_geo, how='inner', on='cell_id')
df_joint3.head()
```

```
#The number of rows and column in the merged dataframe.
df_joint3.shape
```

```
#Concatenating the three merged dataframes.
merged_df = pd.concat([df_joint, df_joint2, df_joint3])
merged_df.head()
```

```
#checking the rows and columns of the the new dataframe.
merged_df.shape
```

```
#Removing any useless columns.
merged_df.drop(['country_a', 'country_b', 'cell_on_site'], axis=1, inplace=True)

merged_df.head()
```

```
#Finding out if there are null or missing vlaues and the sum per column
merged_df.isnull().sum()
```

```
#Deleting the site_id because it has so many missing values which would be problemmatic in analysis
merged_df.drop("site_id", axis=1, inplace=True)
```

```
#Replacing the missing column with a median
merged_df.fillna(np.median, inplace=True)
```

```
#Checking for any other null values
merged_df.isnull().sum()
merged_df.shape
```

```
#Checking datatypes
merged_df.dtypes
```

```
#Changing datatype of datetime column data from object to a datetime type
merged_df.datetime = pd.to_datetime(merged_df.datetime)

# splitting the datetime to date and time

merged_df['date'] = [d.date() for d in merged_df.datetime]

merged_df['time'] = [d.time() for d in merged_df.datetime]

merged_df['date'] = pd.to_datetime(merged_df['date'])

merged_df['time'] = pd.to_datetime(merged_df['time'], format='%H:%M:%S:%f', errors = 'ignore')

# Confirming whether the changes have been made

merged_df.dtypes
```

```
#Cities with the most usage
merged_df.groupby(['date', 'villes']).size()
```

```
#Obtaining data of usage during business hours
import datetime

working_hours =merged_df[(merged_df.datetime.dt.time >= datetime.time(9)) & (merged_df.datetime.dt.time <= datetime.time(17))]

working_hours.head()
```

```
#Determining during working hours which city has the highest usage.
working_hours.groupby('villes')['villes'].count().sort_values(ascending=False)
```

```
#Obtaining data of usage outside business hours
import datetime

off_hours =merged_df[~((merged_df.datetime.dt.time >= datetime.time(9)) & (merged_df.datetime.dt.time <= datetime.time(17)))]

off_hours.head()
```

```
#Cities with the most usage outside business hours
off_hours.groupby('villes')['villes'].count().sort_values(ascending = False)
```

```
#Cities with the most usage in the 3 days
merged_df.groupby('villes')['villes'].count().sort_values(ascending = False)
```
